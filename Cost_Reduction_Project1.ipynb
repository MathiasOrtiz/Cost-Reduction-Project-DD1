# Cost Reduction Project at Dunkin' Donuts

This project aims to reduce costs at Dunkin' Donuts by implementing reusable cups for employees. 
The analysis will compare the current expense of disposable cups, the investment in reusable cups, 
and the time required to recover the investment.

## Defining Variables
The costs of disposable cups, lids, and the number of employees using them daily are established.

```python
medium_cup = 0.1
lid = 0.04
employees_per_day = 12
cups_per_employee = 3
total_cups = employees_per_day * cups_per_employee
total_lids = total_cups
```

## Calculating the Annual Cost of Disposable Cups
This section calculates the cost of using disposable cups and lids over a day, a week, a month, and a year.

```python
cost_cups_per_day = total_cups * medium_cup
cost_lids_per_day = total_lids * lid
total_cost_per_day = cost_cups_per_day + cost_lids_per_day
print('The total cost per day is: ', total_cost_per_day, 'USD')
```

```python
import matplotlib.pyplot as plt
import numpy as np

days = np.array([7, 30, 365])  # Week, Month, Year
costs = total_cost_per_day * days

plt.figure(figsize=(8, 5))
plt.bar(["Week", "Month", "Year"], costs, color=["blue", "green", "red"])
plt.ylabel("Cost in USD")
plt.title("Estimated Cost of Cups and Lids Over Time")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Showing the values
for i, cost in enumerate(costs):
    plt.text(i, cost + 10, f"${cost:.2f}", ha="center", fontsize=12)

# Showing the graph
plt.show()
```

## Implementing Reusable Cups
To reduce costs, reusable cups personalized for each employee are proposed. This section analyzes 
the total investment cost and compares it with the current annual expenditure.


```python
approximate_cost_reusable_cup = 4
employees = 19
total_cost_reusable_cup = approximate_cost_reusable_cup * employees
plt.figure(figsize=(4, 5))
plt.bar('Reusable cups', total_cost_reusable_cup, color='blue')
plt.ylabel('Cost in USD')
plt.title('Cost of Reusable Cups')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
print('The total cost of reusable cups is: ', total_cost_reusable_cup, 'USD')
```

## Investment Recovery Calculation
This section determines how many days it will take to recover the investment in reusable cups compared 
to the current daily spending on disposable cups.

```python
days_to_recover = total_cost_reusable_cup / total_cost_per_day
print('It will take ', days_to_recover, ' days to recover the cost of reusable cups')
```

## Visualization of Investment Recovery
A graph illustrates the accumulated savings over time, highlighting the break-even point 
where the investment in reusable cups equals the savings.


```python
days = np.arange(1, int(days_to_recover) + 10)  # 5 days extra to show the full recovery
savings = days * total_cost_per_day

# Graph
plt.figure(figsize=(8, 5))
plt.plot(days, savings, label="Savings Over Time", color="green", linewidth=2)
plt.axhline(y=total_cost_reusable_cup, color="red", linestyle="--", label=f"Investment Cost ({total_cost_reusable_cup} USD)")
plt.axvline(x=days_to_recover, color="blue", linestyle="--", label=f"Break-even Point ({int(days_to_recover)} days)")

# Labels and title
plt.xlabel("Days")
plt.ylabel("Savings (USD)")
plt.title("Recovery of Investment in Reusable Cups")
plt.legend()
plt.grid(alpha=0.5)

# Showing the graph
plt.show()
```
